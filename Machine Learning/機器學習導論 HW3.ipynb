```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_excel("D:/SCU/Grade3/Concrete_Data.xls")
```

```python
data
```

```python
data.columns = ['A','B','C','D','E','F','G','H','I'] #暫時改名
data
```

```python
a = ["Cement (component 1)(kg in a m^3 mixture)",
     "Blast Furnace Slag (component 2)(kg in a m^3 mixture)",
     "Fly Ash (component 3)(kg in a m^3 mixture)",
     "Water (component 4)(kg in a m^3 mixture)",
     "Superplasticizer (component 5)(kg in a m^3 mixture)",
     "Coarse Aggregate (component 6)(kg in a m^3 mixture)",
     "Fine Aggregate (component 7)(kg in a m^3 mixture)",
     "Age (day)",
     "Concrete compressive strength(MPa, megapascals)"]
```

```python
sc_x = StandardScaler()
test = sc_x.fit_transform(data)
```

```python
data_clean = pd.DataFrame(test,columns =(a)) #資料清整
data_clean
```

```python
data_clean['Concrete compressive strength(MPa, megapascals)'] = data['I']
data_clean
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#散點圖
sns.pairplot(data_clean[a])
plt.tight_layout()
plt.show()
```

```python
#關係矩陣圖
import numpy as np

cm = np.corrcoef(data_clean[a].values.T)

hm = sns.heatmap(cm,
                 cbar = True,
                 annot = True,
                 square = True,
                 fmt = ".2f",
                 annot_kws = {"size" : 10},
                 yticklabels = a,
                 xticklabels = a)
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
data_array = data.values

b = ["Cement (component 1)(kg in a m^3 mixture)",
     "Blast Furnace Slag (component 2)(kg in a m^3 mixture)",
     "Fly Ash (component 3)(kg in a m^3 mixture)",
     "Water (component 4)(kg in a m^3 mixture)",
     "Superplasticizer (component 5)(kg in a m^3 mixture)",
     "Coarse Aggregate (component 6)(kg in a m^3 mixture)",
     "Fine Aggregate (component 7)(kg in a m^3 mixture)",
     "Age (day)"]

data_clean[b]
```

```python
x , y = data_clean[b] , data_clean["Concrete compressive strength(MPa, megapascals)"]
```

```python
len(y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y ,random_state = 8, shuffle=True)

slr = LinearRegression()
slr.fit(x_train,y_train)
print(slr.coef_)
```

```python
y_train_pred = slr.predict(x_train)
y_test_pred = slr.predict(x_test)
```

```python
print("MSE train : %.3f ,test:%.3f" %(mean_squared_error(y_train,y_train_pred) ,mean_squared_error(y_test,y_test_pred)))
print("R^2 train : %.3f ,test:%.3f" %(r2_score(y_train,y_train_pred),r2_score(y_test,y_test_pred)))
print(slr.coef_)

#比較相關係數跟回歸係數，發現：Fly Ash (component 3)(kg in a m^3 mixture)、Coarse Aggregate (component 6)(kg in a m^3 mixture)、Fine Aggregate (component 7)(kg in a m^3 mixture)為異號，所以要刪掉
```

```python
print("MSE train : %.3f ,test:%.3f" %(mean_squared_error(y_train,y_train_pred) ,mean_squared_error(y_test,y_test_pred)))
print("R^2 train : %.3f ,test:%.3f" %(r2_score(y_train,y_train_pred),r2_score(y_test,y_test_pred)))
```
